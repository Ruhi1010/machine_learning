Heat map

```python
diabetes = pd.read_csv('diabetes.csv')
df = pd.DataFrame(diabetes)
```

```python
df.head()
```

```python
correlation = df.corr()
```

```python
# constructing a heatmap
plt.figure(figsize=(10, 10))
ax = sns.heatmap(
    correlation,
    cmap='Blues',
    square=True,
    cbar=True,
    linewidths=0.5,
    linecolor='white'
)
for i in range(correlation.shape[0]):  # loop over rows
    for j in range(correlation.shape[1]):  # loop over columns
        ax.text(
            j + 0.5,                  # x-coordinate (column index + offset for centering)
            i + 0.5,                  # y-coordinate (row index + offset for centering)
            f'{correlation.iloc[i, j]:.2f}',  # formatted text (2 decimal places)
            ha='center',              # horizontal alignment
            va='center',              # vertical alignment
            color='black',            # text color
            fontsize=10               # text size
        )

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dummy correlation data
df = pd.DataFrame(np.random.rand(5, 5), columns=list('ABCDE'))
correlation = df.corr()

# Plot
plt.figure(figsize=(8, 8))
ax = sns.heatmap(
    correlation,
    cmap='Blues',
    square=True,
    cbar=True,
    linewidths=0.5,
    linecolor='white'
)

# Manually add text annotations
for i in range(correlation.shape[0]):
    for j in range(correlation.shape[1]):
        ax.text(j + 0.5, i + 0.5, f'{correlation.iloc[i, j]:.2f}',
                ha='center', va='center', color='black', fontsize=10)

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
